### ✏️ Soru 14: `puan = 65` için not harfini yazdırın:
- 90 ve üstü → `"A"`
- 75 ve üstü → `"B"`
- 60 ve üstü → `"C"`
- 60 altı → `"F"`

```python
puan = 120
try:
 if puan > 100:
    raise ValueError
 elif (puan >= 90):
   print("A")
 elif (puan >= 75):
   print("B")
 elif (puan >= 60):
   print("C")
 elif (puan < 60):
   print("F")
except:
  print("Puan 100'den büyük olamaz!")
```

### ✏️ Soru 24:
Bir alışveriş sepeti uygulaması yapın:

### ✏️ Soru 25:
Bir listede verilen notların ortalamasını hesaplayan `ortalama(notlar)` fonksiyonu yazın.
- `sum()` ve `len()` kullanabilirsiniz
- `[70, 85, 90, 60, 75]` listesiyle test edin

```python
def ortalama(notlar):
    ort = sum(notlar)/len(notlar)
    print(ort)

test_notlari = [70, 85, 90, 60, 75]
print(ortalama(test_notlari))
```
